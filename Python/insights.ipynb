```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("E:\\Data Analysis\\Projects\\Data Analysis End to end\\Data\\cleaned_data.csv")
```

Discount vs Profit - Are high discounts hurting profit?

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='discount', y='profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

It shows that orders with discounts above 30% frequently generate negative profit, indicating that aggressive discounting is hurting margins rather than driving growth. A controlled discount policy or targeted discounts for highâ€‘value customers could help improve overall profit margins.

Revenue & Profit by Category - Which category earns the most and which is actually profitable?

```python
cat = df.groupby('category')[['sales','profit']].sum().reset_index()

cat.plot(x='category', kind='bar', figsize=(8,5))
plt.title("Revenue & Profit by Category")
plt.show()
```

Technology is the top revenueâ€‘generating category and also delivers the highest profit, while Furniture shows comparatively low profit despite strong sales, indicating weaker margins. Focusing on margin improvement in Furniture and expanding highâ€‘performing Technology products could increase overall profitability.

Region Performance - Which region is strong or weak?

```python
region = df.groupby('region')[['sales','profit']].sum().reset_index()

region.plot(x='region', kind='bar', figsize=(8,5))
plt.title("Region Revenue & Profit")
plt.show()
```

The West region contributes the highest profit and revenue, whereas the Central region shows weaker profitability despite moderate sales. Targeted marketing and pricing adjustments in lowerâ€‘margin regions could help balance overall performance.

Repeat vs New Customers - Is growth coming from loyal customers or new ones?

```python
orders_per_customer = df.groupby('customer_id')['order_id'].nunique()
repeat = (orders_per_customer > 1).sum()
new = (orders_per_customer == 1).sum()

print("Repeat Customers:", repeat)
print("New Customers:", new)

print(repeat/(repeat+new)*100.0)
```

Approximately 98% of customers are repeat buyers, indicating extremely strong customer retention. While retention is excellent, increasing new customer acquisition could support longâ€‘term growth and reduce dependency on repeat buyers.
